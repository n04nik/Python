```python
import random
import datetime
# import prettytable                  #   
import matplotlib.pyplot as plt     #   

def BubbleSort(A):                  #  
    for i in range(len(A)):
        for j in range(len(A)-1-i):
            if A[j] > A[j+1]:
                a = A[j]
                A[j] = A[j+1]
                A[j+1] = a


def QuickSort(A, fst, lst):         #  
    if fst >= lst:
        return

    #i, j = fst, lst
    pivot = A[fst]
    # pivot = A[random.randint(fst, lst)]
    first_bigger = fst+1
    while first_bigger <= lst:
        if A[first_bigger] >= pivot:
            break
        first_bigger += 1
    i = first_bigger+1
    while i <= lst:
        if A[i] < pivot:
            A[i], A[first_bigger] = A[first_bigger], A[i]
            first_bigger += 1
        i += 1

    last_smaller = first_bigger-1
    A[fst], A[last_smaller] = A[last_smaller], A[fst]
    QuickSort(A, fst, last_smaller-1)
    QuickSort(A, first_bigger, lst)


def InsertSort(C):
    for i in range(1, len(C)):
        t = C[i]
        j = 1
        while j > 0 and C[j - 1] > t:
            C[j] = C[j-1]
            j -= 1
        C[j] = t

def ShakerSort(D):
    for i in range(0, int(len(D)/2)):
        for j in range(i, len(D) - 1 - i):
            if D[j] > D[j + 1]:
                D[j] = D[j + 1]
        for j in range(len(D)-2-i, i + 1):
            if D[j]<D[j-1]:
                D[j] = D[j - 1]


# table = prettytable.PrettyTable([" ", " ", " "])
x = []
y1 = []
y2 = []
y3 = []
y4 = []


for N in range(1000,5001,1000):
    x.append(N)
    min = 1
    max = N
    A = []
    for i in range (N):
        A.append(int(round(random.random()*(max-min)+min)))

    B = A.copy()

    C = A.copy() 
    print("---")


    D = A.copy()
    print("---")

    t1 = datetime.datetime.now()
    BubbleSort(A)
    t2 = datetime.datetime.now()
    y1.append((t2-t1).total_seconds())
    print("    " + str(N) + "      " + str((t2-t1).total_seconds()) + "c")

    t3 = datetime.datetime.now()
    QuickSort(B, 0, len(B)-1)
    t4 = datetime.datetime.now()
    y2.append((t4 - t3).total_seconds())
    print("   " + str(N) + "      " + str((t4-t3).total_seconds()) + "c")


    t5 = datetime.datetime.now()
    InsertSort(C)
    t6 = datetime.datetime.now()
    y3.append((t6 - t5).total_seconds())
    print("   " + str(N) + "    " + str((t6 - t5).total_seconds()))
    
    t7 = datetime.datetime.now()
    ShakerSort(D)
    t8 = datetime.datetime.now()
    y4.append((t8 - t7).total_seconds())
    print("   " + str(N) + "    " + str((t8 - t7).total_seconds()))



    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}

fig, axs = plt.subplots(3, 1, figsize = (7, 7) ,layout = "constrained", facecolor = 'grey')

axs[0].plot(x, y1)
axs[0].plot(x, y2)
axs[0].plot(x, y3)
axs[0].plot(x, y4)
axs[0].set_title("Sorting Execution Time", fontdict = font1)
axs[0].set_ylabel("Time (s)", fontdict = font2)
axs[0].set_xlabel("Range", fontdict = font2)
axs[0].legend(["Bubble Sort", "Quick Sort", "Insert Sort", "Shaker Sort"])

axs[1].plot(x, y1, 'x')
axs[1].plot(x, y2, 'x')
axs[1].plot(x, y3, 'x')
axs[1].plot(x, y4, 'x')
axs[1].set_title("Sorting Execution Time Dot", fontdict = font1)
axs[1].set_ylabel("Time (s)", fontdict = font2)
axs[1].set_xlabel("Range", fontdict = font2)
axs[1].legend(["Bubble Sort", "Quick Sort", "Insert Sort", "Shaker Sort"])


bar_width = 0.2
ind = range(len(x))

axs[2].bar([i - bar_width * 1.5 for i in ind], y1)
axs[2].bar([i - bar_width / 2 for i in ind], y2)
axs[2].bar([i + bar_width / 2 for i in ind], y3)
axs[2].bar([i + bar_width * 1.5 for i in ind], y4)
axs[2].set_title("Sorting Execution Time Bar", fontdict = font1)
axs[2].set_ylabel("Time (s)", fontdict = font2)
axs[2].set_xlabel("Range", fontdict = font2)
axs[2].legend(["Bubble Sort", "Quick Sort", "Insert Sort", "Shaker Sort"])


#     

fig, ax = plt.subplots(2, 2, figsize = (7, 7) ,layout = "constrained", facecolor = 'grey')

ax[0, 0].plot(x, y1)
ax[0, 0].set_title("Bubble Sort", fontdict = font1)
ax[0, 0].set_ylabel("Time (s)", fontdict = font2)
ax[0, 0].set_xlabel("Range", fontdict = font2)

ax[1, 0].plot(x, y2)
ax[1, 0].set_title("Quick Sort", fontdict = font1)
ax[1, 0].set_ylabel("Time (s)", fontdict = font2)
ax[1, 0].set_xlabel("Range", fontdict = font2)

ax[0, 1].plot(x, y3)
ax[0, 1].set_title("Insert Sort", fontdict = font1)
ax[0, 1].set_ylabel("Time (s)", fontdict = font2)
ax[0, 1].set_xlabel("Range", fontdict = font2)

ax[1, 1].plot(x, y4)
ax[1, 1].set_title("Shaker Sort", fontdict = font1)
ax[1, 1].set_ylabel("Time (s)", fontdict = font2)
ax[1, 1].set_xlabel("Range", fontdict = font2)


plt.show()
```
